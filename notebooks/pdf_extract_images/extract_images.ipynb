```python
from PyPDF2 import PdfReader
```

```python
reader = PdfReader("Downloads/birth_cirtificate.pdf")
```

```python
page_num = 2
page = reader.pages[page_num]

count = 0
for image_file_object in page.images:
    with open("Downloads/" + str(page_num) + str(count) + ".png", "wb") as fp:
        fp.write(image_file_object.data)
        count += 1
print(count)
```
