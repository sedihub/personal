```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("United States of America-2024.csv")

display(df)
```

```python
df["Age_lower"] = df["Age"].apply(lambda x: int(x.split("-")[0]))
df["Age_upper"] = df["Age"].apply(lambda x: int(x.split("-")[1]))

display(df)
display(df.dtypes)
```

```python
population = df[["M", "F"]].sum().to_dict()
print(population)
```

```python
df["M_normalized"] = df["M"] / population["M"]
df["F_normalized"] = df["F"] / population["F"]

display(df)
```

```python
df["M_cumsum"] = df["M"].cumsum()
df["F_cumsum"] = df["F"].cumsum()

display(df)
```

```python
df["M_cumsum_norm"] = df["M_cumsum"] / population["M"]
df["F_cumsum_norm"] = df["F_cumsum"] / population["F"]

display(df)
```

```python
fig = plt.figure(figsize=(12.0, 5.0))
ax = fig.gca()

ax.grid()
ax.plot(df["Age_upper"], df["M_cumsum_norm"], ls="--", label="Male")
ax.plot(df["Age_upper"], df["F_cumsum_norm"], ls="--", label="Female")
ax.legend()

axr = ax.twinx()
axr.plot(df["Age_upper"], df["M_cumsum"], label="Male")
axr.plot(df["Age_upper"], df["F_cumsum"], label="Female")
axr.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(12.0, 5.0))
ax = fig.gca()


ax.grid()
ax.plot(df["Age_upper"], df["M_cumsum_norm"] - df["F_cumsum_norm"])

axr = ax.twinx()
axr.plot(df["Age_upper"], df["M_cumsum"] - df["F_cumsum"])

plt.show()
```
