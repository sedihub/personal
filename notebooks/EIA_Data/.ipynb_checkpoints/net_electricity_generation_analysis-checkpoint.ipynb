```python
import pandas as pd

import numpy as np

from scipy.optimize import curve_fit

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.pyplot import cm
from matplotlib.lines import Line2D
```

```python
df = pd.read_csv("./Net_generation_United_States_all_sectors_monthly.csv", skiprows=[0])

# Split "Month-Year" into "month" and "year"
df[["month", "year"]] = df["Month-Year"].str.split(" ", expand=True)

display(df.columns)
display(df.head(5))
```

```python
columns = ['all fuels (utility-scale) 1000 MWh', 'coal 1000 MWh',
       'natural gas 1000 MWh', 'nuclear 1000 MWh',
       'conventional hydroelectric 1000 MWh', 'wind 1000 MWh',
       'all solar 1000 MWh', 'month']

grouped_df = df.fillna(0.0).groupby("year", as_index=False)[columns].agg(lambda x: list(x))
display(grouped_df.head(5))
```

</br></br></br>

# Annual Trends for Each Source

```python
fig = plt.figure(figsize=(12., 32.0))

months = ["Dec",  "Nov",  "Oct",  "Sep",  "Aug",  "Jul",  "Jun",  "May",  "Apr",  "Mar",  "Feb",  "Jan"]
months.reverse()

columns = ["coal 1000 MWh", "natural gas 1000 MWh", "nuclear 1000 MWh",
    "conventional hydroelectric 1000 MWh", "wind 1000 MWh",
    "all solar 1000 MWh"]

num_years = len(grouped_df.index)
colors = cm.rainbow(np.linspace(0, 1, num_years))

for n, source in enumerate(columns, start=1):
    plt.subplot(len(columns), 1, n)
    ax = plt.gca()
    ax.grid(lw=0.5, color="darkgrey")
    ax.set_ylabel(source.title().replace("1000 Mwh", "[TWh]"), color="darkgrey", 
                  fontsize=12, fontweight="bold")
    for idx, row in grouped_df.iterrows():
        ax.plot(np.arange(len(row["month"])), 0.001 * np.array(row[source][::-1]), 
                lw=1.5, alpha=0.75, color=colors[idx], label=row["year"])
    ax.set_xticks(np.arange(12))
    if n < len(columns):
        ax.set_xticklabels([])
    else:
        ax.set_xticklabels(months, va="top", ha="left", rotation=-45, color="darkgrey", fontsize=12)
    ax.set_ylim(-0.5,)
plt.legend(loc="lower center", bbox_to_anchor=(1.05, -0.025),
            fancybox=True, shadow=True, ncol=1)
plt.show()
```

```python
fig = plt.figure(figsize=(12., 48.0))

months = ["Dec",  "Nov",  "Oct",  "Sep",  "Aug",  "Jul",  "Jun",  "May",  "Apr",  "Mar",  "Feb",  "Jan"]
months.reverse()


columns = ["coal 1000 MWh", "natural gas 1000 MWh", "nuclear 1000 MWh",
    "conventional hydroelectric 1000 MWh", "wind 1000 MWh",
    "all solar 1000 MWh"]

num_years = len(grouped_df.index)
num_cols = 2
num_rows = (num_years // num_cols + int(num_years % num_cols != 0))
gs = gridspec.GridSpec(ncols=num_cols, nrows=num_rows, 
                       figure=fig, wspace=0.025, hspace=0.15)
colors = cm.plasma(np.linspace(0, 1, len(columns)))
colors = ["darkgrey", "royalblue", "lightgreen", "navy", "orange", "red"]

for idx, row in grouped_df.iterrows():
    row_idx = idx // num_cols
    col_idx = idx % num_cols
    ax = fig.add_subplot(gs[row_idx, col_idx])
    ax.grid(lw=0.5, color="darkgrey")
    cumulative = []
    ax.set_xlim(0.0, 425.0)
    ax.set_title(row["year"], color="darkgrey", fontsize=12, fontweight="bold")
    for src_idx, source in enumerate(columns):
        if cumulative == []:
            left = 0.0
        else:
            left = np.stack(cumulative, axis=0).sum(axis=0)
        x = 0.001 * np.array(row[source][::-1])
        cumulative.append(x)
        ax.barh(y=np.arange(len(row["month"])), width=x,
                left=left, height=0.8, align="center",# edgecolor="black",
                alpha=0.75, color=colors[src_idx],
                label=source.title().replace("1000 Mwh", ""))
    ax.set_yticks(np.arange(12))
    ax.set_axisbelow(True)
    
    if row_idx != (num_rows - 1):
        ax.set_xticklabels([])
    else:
        ax.set_xlabel("[TWh]", color="darkgrey", fontsize=12, fontweight="bold")
    
    if num_cols > 1 and col_idx > 0:
        ax.set_yticklabels([])
    else:
        ax.set_yticklabels(months, va="center", ha="right", rotation=0.0, color="darkgrey", fontsize=8)
        
    # ax.legend(loc="lower center", bbox_to_anchor=(0.5, -0.05),
    #     fancybox=True, shadow=True, ncol=5)
plt.legend(loc="lower center", bbox_to_anchor=(0.0, -0.35),
            fancybox=True, shadow=True, ncol=6)
plt.show()
```

```python
fig = plt.figure(figsize=(12.0, 24.0))

months = ["Dec",  "Nov",  "Oct",  "Sep",  "Aug",  "Jul",  "Jun",  "May",  "Apr",  "Mar",  "Feb",  "Jan"]
columns = ["coal 1000 MWh", "natural gas 1000 MWh", "nuclear 1000 MWh",
    "conventional hydroelectric 1000 MWh", "wind 1000 MWh",
    "all solar 1000 MWh"]

num_years = len(grouped_df.index)
num_cols = 4
num_rows = (num_years // num_cols + int(num_years % num_cols != 0))
gs = gridspec.GridSpec(ncols=num_cols, nrows=num_rows, 
                       figure=fig, wspace=0.025, hspace=0.025)

colors = cm.plasma(np.linspace(0, 1, len(columns)))
# colors = ["darkgrey", "royalblue", "lightgreen", "navy", "orange", "red"]

for idx, row in grouped_df.iterrows():
    row_idx = idx // num_cols
    col_idx = idx % num_cols
    ax = fig.add_subplot(gs[row_idx, col_idx])
    ax.set_title(row["year"], color="black", fontsize=10, fontweight="bold")
    ax.set(aspect="equal")
    year_data_array = np.array(row[columns].to_list())
    months = row["month"]
    for month_idx, month in enumerate(months):
        month_gen_elec_twh = 0.001 * year_data_array[:, month_idx]
        scale_radius = 0.3
        radius = scale_radius * (month_idx + 1) / len(month)
        width = scale_radius * 1.0 / len(month)
        ax.pie(month_gen_elec_twh, radius=radius, 
               colors=colors, wedgeprops=dict(width=width, edgecolor="white", linewidth=0.5),
              )#labels=[x.title().replace("1000 Mwh", "") for x in columns])
        
# Custom legend:
legend_text = []
custom_lines = []
for idx, eng_src in enumerate(columns):
    legend_text.append(eng_src.title().replace("1000 Mwh", ""))
    custom_lines.append(Line2D([0], [0], color=colors[idx], lw=6))
plt.legend(custom_lines, legend_text, title="Electricity Generation Source",
           bbox_to_anchor=(1.2, 0.9), fancybox=True, shadow=True, ncol=1)
        
plt.show()
```
